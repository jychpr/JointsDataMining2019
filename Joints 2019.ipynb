<h3>Buat misah input sama output data train </h3>

```python
X, y = df.iloc[:,1:-1], df.iloc[:,-1]
X.head()
y
```

<h3>Ini buat ganti kategori jadi numerik, 0 = campur, 1 = putra, 2 = putri </j3>

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(y)
label_encoded_y = label_encoder.transform(y)

y = pd.DataFrame({'Gender':label_encoded_y.ravel()})
```

```python
from sklearn.impute import SimpleImputer

X_fac, X_poi_count = X.iloc[:,[0,1,2,3,4,5,7]], X.iloc[:,8:]

#imputasi nilai numerik dengan mediannya
imp = SimpleImputer(missing_values=np.nan,strategy='median')
Xt_poi_count =  imp.fit_transform(X_poi_count)

#imputasi nilai kategori dengan nilai 0 - fitur fac
imp2 = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=0)
Xt_fac = imp2.fit_transform(X_fac)

# from sklearn.preprocessing import Normalizer
# normalizer = Normalizer().fit(Xt_poi_count)
# Xt_poi_count = normalizer.transform(Xt_poi_count)

Xt = np.hstack((Xt_fac, Xt_poi_count))
```

<h2>Coba-coba model sama gonta-ganti hyper parameter tiap model diukur pakai 10th-fold cross validation.</h2>
*harusnya ada pake algoritme hyperparameter tuning, tapi lupa

```python
from xgboost import XGBClassifier


model = XGBClassifier(random_state=30, learning_rate=0.1, booster='dart',max_depth=7, colsample_bytree=0.6, n_jobs=-1)

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
kfold = StratifiedKFold(n_splits=10, random_state=7)
results = cross_val_score(model, Xt, y.values.ravel(), cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))

# model.fit(Xt, y.values.ravel())
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=20, n_estimators=100, max_depth=19, n_jobs=-1, min_samples_leaf = 4,
                              min_samples_split= 10)

kfold = StratifiedKFold(n_splits=10, random_state=7)
results = cross_val_score(model, Xt, y.values.ravel(), cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
model.fit(Xt, y.values.ravel())
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=20, max_depth=5,min_samples_split=7, min_samples_leaf=0.1)

kfold = StratifiedKFold(n_splits=10, random_state=7)
results = cross_val_score(model, Xt, y.values.ravel(), cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
from sklearn.neighbors import KNeighborsClassifier

c = [2,3,5,6,20,300,500,1000,2000,10000]
for a in c :
    model = KNeighborsClassifier(n_neighbors=a,algorithm='auto',p=2, weights='distance')

    kfold = StratifiedKFold(n_splits=10, random_state=7)
    results = cross_val_score(model, Xt, y.values.ravel(), cv=kfold, error_score='raise')
    print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
df_test2 = df_test.drop(columns='fac_7')
X_test_fac = df_test2.iloc[:,1:8]

X_test_poi = df_test2.iloc[:,8:]
X_test_poi = (X_test_poi)

Xt_test = np.hstack((X_test_fac, X_test_poi))
Xt_test
```

```python
y_pred = model.predict(Xt_test)
predictions = [round(value) for value in y_pred]
file = [int(value['id']) for i,value in df_test.iterrows()]
```
